```python
import pandas as pd

df = pd.read_csv('aerofit_treadmill_data.csv')
df
```

```python
pd.crosstab(df['Product'], df['Gender'], normalize='index').round(3)*100
```

> based on gender, KP781 is heavily skewed to Males;
> KP281 & KP 481 (cheaper versions) is equally popular

```python
pd.crosstab(df['Product'], df['MaritalStatus'], normalize='index').round(3)*100
```

> Partnered people are more likely to purchaese any product, as compared to singles

```python
pd.crosstab(df['Product'], df['Fitness'], normalize='index').round(3)*100
```

> fitness level 3 people prefer cheaper versions - KP281, KP481 ; whereas the most expensive verioin is only popular among fitness level 5 people

```python
df['Age_Group'] = pd.cut(df['Age'], bins=range(15, 65, 5), right=False, labels=[f'{i}-{i+4}' for i in range(15, 60, 5)])
print(df[['Age', 'Age_Group']].head())

pd.crosstab(df['Product'], df['Age_Group'], normalize='index').round(3)*100
```

> age groups 20-24, 25-29 are most avid purchasers; followed by age 30-34 ones

```python
pd.crosstab(df['Product'], df['Education'], normalize='index').round(3)*100
```

> people with 14 years of education, are more likely to buy cheaper products KP281, KP481; people with 16 years of eduction are best targets for any type; and with 18 years of education are best for expenseive verisns KP781; beyond 18 not much use of targeting

```python
pd.crosstab(df['Product'], df['Usage'], normalize='index').round(3)*100
```

people with anticipated usage of 2-4 times a wekk are more likely to buy cheaper versions - KP281, KP481. whereas people who buy the expensive verison KP781 are mostly the ones who anticipate using 4-6 times a week

```python
print(df['Income'].describe())
```

```python
import numpy as np

# Based on the describe output, let's create custom bins
bins = [df['Income'].min(), 30000, 40000, 50000, 60000, 70000, 80000, df['Income'].max()]
labels = ['<30k', '30k-40k', '40k-50k', '50k-60k', '60k-70k', '70k-80k', '>80k']

df['Income_Group'] = pd.cut(df['Income'], bins=bins, labels=labels, right=False)
print(df[['Income', 'Income_Group']].head())
```

```python
pd.crosstab(df['Product'], df['Income_Group'], normalize='index').round(3)*100
```

people with incomes between 30k to 60k are more likely to spend on cheaper versions - KP281, KP481. whereas people with incomes 60k and abvoe are more likely to buy the expensive vrsion KP781

```python
print(df['Miles'].describe())
```

```python
# Based on the describe output, let's create custom bins for 'Miles'
bins = [df['Miles'].min(), 50, 100, 150, 200, df['Miles'].max() + 1] # Add +1 to max to ensure max value is included
labels = ['<50', '50-99', '100-149', '150-199', '>=200']

df['Miles_Group'] = pd.cut(df['Miles'], bins=bins, labels=labels, right=False)
print(df[['Miles', 'Miles_Group']].head())
```

```python
pd.crosstab(df['Product'], df['Miles_Group'], normalize='index').round(3)*100
```
