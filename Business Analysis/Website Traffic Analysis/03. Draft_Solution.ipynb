```python
df[df['event']=='pageview']\
  .groupby('country')\
  .agg(n_views = ('event','count'))\
  .sort_values(by='n_views', ascending=False)
```

```python
print('click_rate', df[df['event']=='click'].shape[0]/df[df['event']=='pageview'].shape[0])
```

```python
df1 = df[df['event']=='click'].groupby('linkid').agg(n_link_clicks = ('event','count'))
df2 = df[df['event']=='pageview'].groupby('linkid').agg(n_link_pageviews = ('event','count'))

df3 = pd.merge(df1, df2, how='inner', on='linkid')
df3['click_rate'] = df3['n_link_clicks']/df3['n_link_pageviews']
df3.sort_values(by='click_rate', ascending=False)
```

```python
df4 = df[df['event']=='click'].groupby('linkid').agg(n_link_clicks = ('event','count'))
df5 = df[df['event']=='preview'].groupby('linkid').agg(n_link_previews = ('event','count'))

df6 = pd.merge(df4, df5, how='inner', on='linkid')


from scipy.stats import pearsonr

r, p_value = pearsonr(
    df6["n_link_clicks"],
    df6["n_link_previews"]
)

print("Pearson r:", r)
print("p-value:", p_value)


from scipy.stats import spearmanr

rho, p_value_spearman = spearmanr(
    df6["n_link_clicks"],
    df6["n_link_previews"]
)

print("\nSpearman rho:", rho)
print("p-value:", p_value_spearman)
```
